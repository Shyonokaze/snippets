# Process X-ray EDS map from qm-04093-R138t

Process the EDS map images exported from the Oxford AZtec software from specimen qm-04093-R138t-l3-sm-map-al. The map was acquired on 2014-05-14 at 3:2p pm. The ROI was `96x181` pixels and was recorded at 7 kV and 1750X with a working distance of 5 mm. Under these conditions the original scale in nm/px was

```python
nmPerPx = round(1000.*14.5/181.0,2)
print(nmPerPx)
```

Let's load the packages we need

```python
import numpy as np
import os
from skimage import io, exposure, img_as_uint, img_as_float, img_as_ubyte,color, transform
io.use_plugin('freeimage')
```

We will scale the images by a magnification factor of

```python
mag = 2.0
```

```python
edsRoot = os.environ['EDS_ROOT']
relDir = '/Oxford/QM14-04-02A-Armstrong/reports/R138t-l3-roi/'
imgDir = edsRoot + relDir
print(imgDir)
```

```python
im1 = io.imread(imgDir+'1-CK.png')
im2 = io.imread(imgDir+'2-OK.png')
im3 = io.imread(imgDir+'3-AgL.png')
im4 = io.imread(imgDir+'4-line.png')
```

Note that the AZtec software scaled the ROI image by a scale factor (sf) of

```python
sf = float(im4.shape[1])/float(im1.shape[1])
print(sf)
h = float(im1.shape[0])*sf
print(int(h))
```

First we need to crop the scale bar off the ROI image. Then we need to map the floating point gray level image to 0 to 255 gray levels. Then we can take a peek at the cropped ROI. 

```python
%pylab inline
im4a = im4[0:h-1,]
maxV = amax(im4a.data)
minV = amin(im4a.data)
im4b = (1.0/(maxV-minV))*(im4a.data-minV)
print(amin(im4b.data))
print(amax(im4b.data))
im4b = img_as_ubyte(im4b,force_copy=False)
imshow(im4b, cmap=plt.cm.gray)
print()
print(amax(im4b.data))
print(im4b.dtype)
```

Now we can scale this image by our desired magnification factor and convert it to an RGB image.

```python
im4c=transform.rescale(im4b, (mag,mag), order=2)
imshow(im4c, cmap=plt.cm.gray)
im4d = color.gray2rgb(im4c)
imshow(im4d, cmap=plt.cm.gray)
```

Now let's scale the three X-ray map images

```python
im1a=transform.rescale(im1, (sf*mag,sf*mag), order=3)
imshow(im1a)
```

```python
im2a=transform.rescale(im2, (sf*mag,sf*mag), order=3)
imshow(im2a)
```

```python
im3a=transform.rescale(im3, (sf*mag,sf*mag), order=3)
imshow(im3a)
```

Now let's make an image to hold the stitched images

```python
im5 = transform.rescale(im3a, (2,2), order=3)
```

Let's fill the image with the components

```python
h = im3a.shape[0]
w = im3a.shape[1]
im5[0:h,0:w, :] = im1a
im5[h:2*h,0:w, :] = im2a
im5[0:h,w:2*w, :] = im3a
im5[h:2*h-4,w:2*w, :] = im4d[:,:,0:3]
imshow(im5)
#io.imsave("R138-xray-map.png",im5)
print(im5.shape)
import scipy
scipy.misc.imsave('R138-xray-map.png', im5)
```

Note that this final image has a scale factor of

```python
finNmPerPx = round(nmPerPx /(sf * mag), 2)
print(finNmPerPx)
```
