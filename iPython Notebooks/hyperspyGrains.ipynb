# Analyze clumped grains

Use the I/O capabilities of hyperspy and scikits-image to analyze clumped grains

```python
import numpy as np
from scipy import ndimage
from skimage.morphology import watershed
from skimage.filter import threshold_otsu
from skimage.feature import peak_local_max
```

```python
im = load('sm-grains.tif')
```

```python
#im.plot()
```

```python
dat = im.data
print(dat.shape)
print(dat.dtype)
```

```python
%pylab inline
plt.imshow(dat)
```

```python
thr = threshold_otsu(dat)
print(thr)
thrImg = dat <= thr
plt.imshow(thrImg)
```

```python
distance = ndimage.distance_transform_edt(thrImg)
plt.imshow(distance)
```

```python
local_maxi = peak_local_max(distance, indices=False, footprint=np.ones((3, 3)), labels=thr)
markers = ndimage.label(local_maxi)[0]
labels = watershed(-distance, markers, mask=dat)
```

```python
plt.imshow(labels)
```
