# Test freeimage

On MacOSX install freeimage with homebrew ``brew install freeimage``. This was inspired by

http://stackoverflow.com/questions/24240039/save-numpy-array-as-image-with-high-precision-16-bits-with-scikit-image

```python
import numpy as np
from skimage import io, exposure, img_as_uint, img_as_float

io.use_plugin('freeimage')
```

```python
im = np.array([[1., 2.], [3., 4.]], dtype='float64')
im = exposure.rescale_intensity(im, out_range='float')
im = img_as_uint(im)

io.imsave('test_16bit.png', im)
im2 = io.imread('test_16bit.png')
```

```python
im2
```

As for 3D arrays, you need to construct the array properly and then it'll work:

```python
im = np.linspace(0, 1., 300).reshape(10, 10, 3)
im = exposure.rescale_intensity(im, out_range='float')
im = img_as_uint(im)

io.imsave('test_16bit.png', im)
im2 = io.imread('test_16bit.png')
```

```python
%pylab inline
imshow(im)
```
