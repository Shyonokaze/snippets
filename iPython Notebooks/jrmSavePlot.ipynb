# Saving a Matplotlib figure as a png

I have a frequent need to plot results and save figures as png or pdf files to include in a $\LaTeX$ report.

First setup to use numpy and plot results inline.

```python
import numpy as np
%pylab inline
```

```python
x = np.linspace(0, 2*np.pi, 300)
y = np.sin(x)
y2 = np.sin(x**2)
dpi = 72
xpixels, ypixels = (640, 480)
xinch = float(xpixels)/float(dpi)
yinch = float(ypixels)/float(dpi)
plt.figure(figsize=(xinch,yinch))
plt.plot(x, y, label=r'$\sin(x)$')
plt.plot(x, y2, label=r'$\sin(x^2)$')
plt.title('Some functions')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
f = plt.gcf()
f.savefig('test.png', dpi=dpi, transparent=True)
f.savefig('test.pdf')
```
