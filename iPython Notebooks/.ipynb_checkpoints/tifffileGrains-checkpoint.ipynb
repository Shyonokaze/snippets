# Analyze clumped grains

Use the I/O capabilities of Christof Gohlke's tifffile and scikits-image to analyze clumped grains. This is a work in progress... First, we load what we need.

```python
import numpy as np
import tifffile
from scipy import ndimage
from skimage.morphology import watershed
from skimage.filter import threshold_otsu
from skimage.feature import peak_local_max
```

Now we load the 16 bit/pix ``.tif`` image output from analySIS after opening the DigitalMicrograph ``.dm3`` file. ``tifffile`` lets us convert the image to an ``numpy`` array.

```python
tif = tifffile.TiffFile('grains.tif')
dat = tif[0].asarray()
```

Let's get some basic information about the image

```python
print(type(dat))
print(dat.dtype)
print(dat.shape)
print(np.amin(dat), np.amax(dat))
```

Now let's compute a gray level threshold

```python
thr = threshold_otsu(dat)
print(thr)
thrImg = dat <= thr
```

And let's compute the Euclidean Distance Map for our binarized imae

```python
distance = ndimage.distance_transform_edt(thrImg)
```

Now let's take a peek at our original and thresholded images and distance map

```python
%pylab inline
fig, axes = plt.subplots(ncols=3, figsize=(9, 3))
ax0, ax1, ax2 = axes
ax0.imshow(dat, cmap=plt.cm.gray, interpolation='nearest')
ax0.set_title('Original')
ax1.imshow(thrImg, cmap=plt.cm.gray, interpolation='nearest')
ax1.set_title('Binary')
ax2.imshow(distance, cmap=plt.cm.Spectral, interpolation='nearest')
ax2.set_title('EDM')
for ax in axes:
  ax.axis('off')
plt.subplots_adjust(hspace=0.01, wspace=0.01, top=1, bottom=0, left=0, right=1)
```

```python
local_maxi = peak_local_max(distance, indices=False, footprint=np.ones((3, 3)), labels=thr)
markers = ndimage.label(local_maxi)[0]
labels = watershed(-distance, markers, mask=dat)
```

```python
plt.imshow(labels)
```
